```
import os
import glob
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

```

```
def isbright(image, dim=10, thresh=0.5):
    # Resize image to 10x10
    image = cv2.resize(image, (dim, dim))
    # Convert color space to LAB format and extract L channel
    L, A, B = cv2.split(cv2.cvtColor(image, cv2.COLOR_BGR2LAB))
    # Normalize L channel by dividing all pixel values with maximum pixel value
    #L = L/np.max(L)
    # Return True if mean is greater than thresh else False
    npm = np.mean(L)
    print(npm)
    return npm > thresh

```

```
# load image freom disk
print("new")
image1 = cv2.imread("PhillipsH4RacingVisionVsHalonixBrightnessComparison/new.jpg")
image = image1
# find if image is bright or dark
text = "bright" if isbright(image) else "dark"

print("old")
image2 = cv2.imread("PhillipsH4RacingVisionVsHalonixBrightnessComparison/old.jpg")
image = image2

# find if image is bright or dark
text = "bright" if isbright(image) else "dark"


# write text on image
#cv2.putText(image, "{}".format(text), (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 0, 255), 2)

# show image
plt.figure(figsize=(10,10))
plt.imshow(cv2.cvtColor(image2, cv2.COLOR_BGR2RGB))
plt.show()

plt.figure(figsize=(10,10))
plt.imshow(cv2.cvtColor(image2, cv2.COLOR_BGR2RGB))
plt.show()
```
